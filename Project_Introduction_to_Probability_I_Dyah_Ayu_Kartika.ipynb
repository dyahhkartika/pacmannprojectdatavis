# Analisa Variabel Kategorik

```python
#1
print('1. Gender mana yang memiliki tagihan paling tinggi?')
charges = df.groupby(['sex'])['charges'].sum()
sns.barplot(y = charges, x = ['Female', 'Male']);
plt.title('Charges by Gender')
plt.ylabel('Charges');
plt.show()
print('Laki-laki memiliki tagihan lebih tinggi dibanding perempuan.')


#2
print('\n\n\n2. Distribusi peluang tagihan di tiap-tiap region:')
charges = df.groupby(['region'])['charges'].sum() / df['charges'].sum()
sns.barplot(y = charges, x = ['southwest','southeast','northwest','northeast']);
plt.title('Proportion of Charges in Each Region')
plt.ylabel('Proportion');
plt.show()


#3
print('\n\n\n3. Apakah setiap region memiliki proporsi data banyak orang yang sama?')
people = df.groupby(['region'])['sex'].count() / len(df)
sns.barplot(y = people, x = ['southwest','southeast','northwest','northeast']);
plt.title('Proportion of People in Each Region')
plt.ylabel('Proportion');
plt.show()
print('Region northwest memiliki proporsi user tertinggi jika dibandingkan dengan region lainnya.')


#4
print('\n\n\n4. Mana yang lebih tinggi proporsi perokok atau non perokok?')
propsmoker = df.groupby(['smoker'])['smoker'].count() / len(df)
sns.barplot(y = propsmoker, x = ['Smokers', 'Non-Smokers']);
plt.title('Proportion of Smokers and Non-Smokers')
plt.ylabel('Proportion');
plt.show()
print('Proporsi perokok jauh lebih tinggi jika dibandingkan dengan proporsi non-perokok.')
```

# Analisa Variabel Kontinu

```python
#1
condition = (df['bmi']>25) & (df['charges']>16700)
conditional_price = df['bmi']>25
conditionnn1 = df[condition]['charges'].count() / df[conditional_price]['charges'].count()
print('1. Mencari kemungkinan terjadi, seorang perokok dengan BMI di atas 25 akan mendapatkan tagihan kesehatan di atas 16.700: ', conditionnn1)


#2
condition2 = (df.smoker == 'yes') & (df['charges']>16700)
conditiontambah = df.smoker == 'yes'
conditionnn = df[condition2]['charges'].count() / df[conditiontambah]['charges'].count()
print('\n\n\n2. Berapa peluang seseorang acak tagihan kesehatannya di atas 16.7k diketahui dia adalah perokok: ', conditionnn)


#3
condition3 = (df['bmi']<25) & (df['charges']>16700)
conditional3_price = df['bmi']>25
conditionnn11 = df[condition3]['charges'].count() / df[conditional3_price]['charges'].count()
print('\n\n\n3. Mana yang lebih mungkin terjadi: ', '\na. Seseorang dengan BMI di atas 25 mendapatkan tagihan kesehatan di atas 16.7k, atau \nb. Seseorang dengan BMI di bawah 25 mendapatkan tagihan kesehatan di atas 16.7k')
print('Probabilitas a adalah ', conditionnn1, 'dan probabilitas b adalah ', conditionnn11)
if conditionnn1 > conditionnn11:
  print('Kejadian a lebih mungkin terjadi.')
else:
  print('Kejadian b lebih mungkin terjadi.')


#4
condition4 = (df.smoker == 'no') & (df['charges']>16700) & (df['bmi']>25)
conditional4_price = (df.smoker == 'no') & (df['bmi']>25)
conditionnn111 = df[condition4]['charges'].count() / df[conditional4_price]['charges'].count()
condition5 = (df.smoker == 'yes') & (df['charges']>16700) & (df['bmi']>25)
conditional5_price = (df.smoker == 'yes') & (df['bmi']>25)
conditionnn1111 = df[condition5]['charges'].count() / df[conditional5_price]['charges'].count()
print('\n\n\n4. Mana yang lebih mungkin terjadi: ', '\na. Seseorang perokok dengan BMI di atas 25 mendapatkan tagihan kesehatan di atas 16.7k, atau \nb. Seseorang non perokok dengan BMI di atas 25 mendapatkan tagihan kesehatan di atas 16.7k')
print('Probabilitas a adalah ', conditionnn111, 'dan probabilitas b adalah ', conditionnn1111)
if conditionnn111 > conditionnn1111:
  print('Kejadian b lebih mungkin terjadi.')
else:
  print('Kejadian a lebih mungkin terjadi.')
```

# Pengujian Hipotesis

```python
import scipy.stats as stats
import math
from statsmodels.stats.proportion import proportions_ztest
from statsmodels.stats.weightstats import ztest as ztest
from scipy.stats import ttest_ind

#1
print('1. Tagihan kesehatan perokok lebih tinggi daripada tagihan kesehatan non perokok.')
print('Uji Hipotesis:', '\nH0 = charges smoker <= charges non smoker.', '\nH1 = charges smoker > non smoker.')
alpha = 0.05
print('alpha = ', alpha)
charges_smoker = df[(df.smoker == 'yes')]['charges']
charges_nonsmoker = df[(df.smoker == 'no')]['charges']
ax0 = sns.distplot(charges_smoker, color='tab:blue')
ax0.axvline(np.mean(charges_smoker),color='red',linestyle='--', label='mean')
ax0.axvline(np.median(charges_smoker),color='green',linestyle='--', label='median')
plt.title('Charges Smoker')
plt.legend()
plt.show()
ax1 = sns.distplot(charges_nonsmoker, color='tab:pink')
ax1.axvline(np.mean(charges_nonsmoker),color='red',linestyle='--', label='mean')
ax1.axvline(np.median(charges_nonsmoker),color='green',linestyle='--', label='median')
plt.title('Charges Non Smoker')
plt.legend()
plt.show()
print('Skew of smoker:          %.4f' % stats.skew(charges_smoker))
print('Skew of non smoker:      %.4f' % stats.skew(charges_nonsmoker))
print('Kurtosis of smoker:     %.4f' % stats.kurtosis(charges_smoker))
print('Kurtosis of non smoker:  %.4f' % stats.kurtosis(charges_nonsmoker))
print('Dari uji tersebut terlihat bahwa distribusi charges non smoker skew positif dan lepto kurtosis sehingga akan dilakukan transformasi box-cox.')
charges_smoker1, _ = stats.boxcox(charges_smoker) 
charges_nonsmoker1, _ = stats.boxcox(charges_nonsmoker) 
ax2 = sns.distplot(charges_smoker1, color='tab:blue')
ax2.axvline(np.mean(charges_smoker1),color='red',linestyle='--', label='mean')
ax2.axvline(np.median(charges_smoker1),color='green',linestyle='--', label='median')
plt.title('Charges Smoker Box Cox')
plt.legend()
plt.show()
ax3 = sns.distplot(charges_nonsmoker1, color='tab:pink')
ax3.axvline(np.mean(charges_nonsmoker1),color='red',linestyle='--', label='mean')
ax3.axvline(np.median(charges_nonsmoker1),color='green',linestyle='--', label='median')
plt.title('Charges Non Smoker Box Cox')
plt.legend()
plt.show()
print('Skew of smoker:          %.4f' % stats.skew(charges_smoker1))
print('Skew of non smoker:      %.4f' % stats.skew(charges_nonsmoker1))
print('Kurtosis of smoker:     %.4f' % stats.kurtosis(charges_smoker1))
print('Kurtosis of non smoker:  %.4f' % stats.kurtosis(charges_nonsmoker1))
print('Setelah dilakukan transformasi box cox terlihat bahwa masing-masing distribusi mendekati distribusi normal.')
statistic1, pval1 = stats.shapiro(charges_smoker1)
statistic2, pval2 = stats.shapiro(charges_nonsmoker1)
astat1, crit1, sig1 = stats.anderson(charges_smoker1, dist='norm')
astat2, crit2, sig2 = stats.anderson(charges_nonsmoker1, dist='norm')
#print results
blah1 = astat1/crit1[4]
blah2 = astat2/crit2[4]
print('Shapiro-Wilks test')
print('Charges smoker:      p-value: %.8f' % pval1)
print('Charges non smoker:  p-value: %.8f' % pval2)
print('\nAnderson-Darling Test')
print('Charges smoker:      Anderson-Darling statistic is %.2f times the critical value for alpha = 0.01' % blah1)
print('Charges non smoker:  Anderson-Darling statistic is %.2f times the critical value for alpha = 0.01' % blah2)
print('\nNumber of samples:')
print('Charges smoker:      n = {}' .format(len(charges_smoker)))
print('Charges non smoker:  n = {}' .format(len(charges_nonsmoker)))
stat, pval = stats.levene(charges_smoker1, charges_nonsmoker1, center='mean')
if pval < 0.05:
    print('Tolak H0: Varians mempunyai nilai yang berbeda.\np-value: %.8f' % pval)
else:
    print('Gagal tolak H0: H0 bernilai sama di semua sampel.\np-value: %.8f' % pval)
statistic, pval = stats.ttest_ind(charges_smoker1, charges_nonsmoker1, equal_var = False) 
if pval1 < 0.05:
    print('Tolak H0: Charges smoker lebih tinggi dibanding charges non smoker.\np-value: %.8f' % pval1)
else:
    print('Gagal tolak H0: Charges smoker dan charges non smoker bernilai sama.\np-value: %.8f' % pval1)


#2
print('\n\n\n2. Tagihan kesehatan dengan BMI di atas 25 lebih tinggi daripada tagihan kesehatan dengan BMI di bawah 25.')
print('Uji Hipotesis:', '\nH0 = charges BMI di atas 25 <= charges BMI di bawah 25.', '\nH1 = charges BMI di atas 25 > BMI di bawah 25.')
bmiatas25 = df[(df['bmi']>25)]['charges']
bmibawah25 = df[(df['bmi']<25)]['charges']
ax4 = sns.distplot(bmiatas25, color='tab:blue')
ax4.axvline(np.mean(bmiatas25),color='red',linestyle='--', label='mean')
ax4.axvline(np.median(bmiatas25),color='green',linestyle='--', label='median')
plt.title('Charges BMI di Atas 25')
plt.legend()
plt.show()
ax5 = sns.distplot(bmibawah25, color='tab:pink')
ax5.axvline(np.mean(bmibawah25),color='red',linestyle='--', label='mean')
ax5.axvline(np.median(bmibawah25),color='green',linestyle='--', label='median')
plt.title('Charges BMI di Bawah 25')
plt.legend()
plt.show()
print('Skew of BMI atas 25:          %.4f' % stats.skew(bmiatas25))
print('Skew of BMI bawah 25:      %.4f' % stats.skew(bmibawah25))
print('Kurtosis of BMI atas 25:     %.4f' % stats.kurtosis(bmiatas25))
print('Kurtosis of BMI bawah 25:  %.4f' % stats.kurtosis(bmibawah25))
print('Dari uji tersebut terlihat bahwa distribusi charges BMI atas 25 skew positif sehingga akan dilakukan transformasi box-cox.')
bmiatas251, _ = stats.boxcox(bmiatas25) 
bmibawah251, _ = stats.boxcox(bmibawah25) 
ax6 = sns.distplot(bmiatas251, color='tab:blue')
ax6.axvline(np.mean(bmiatas251),color='red',linestyle='--', label='mean')
ax6.axvline(np.median(bmiatas251),color='green',linestyle='--', label='median')
plt.title('Charges BMI di Atas 25')
plt.legend()
plt.show()
ax7 = sns.distplot(bmibawah251, color='tab:pink')
ax7.axvline(np.mean(bmibawah251),color='red',linestyle='--', label='mean')
ax7.axvline(np.median(bmibawah251),color='green',linestyle='--', label='median')
plt.title('Charges BMI di Bawah 25')
plt.legend()
plt.show()
print('Skew of smoker:          %.4f' % stats.skew(bmiatas251))
print('Skew of non smoker:      %.4f' % stats.skew(bmibawah251))
print('Kurtosis of smoker:     %.4f' % stats.kurtosis(bmiatas251))
print('Kurtosis of non smoker:  %.4f' % stats.kurtosis(bmibawah251))
print('Setelah dilakukan transformasi box cox terlihat bahwa masing-masing distribusi mendekati distribusi normal.')
statistic3, pval3 = stats.shapiro(bmiatas251)
statistic4, pval4 = stats.shapiro(bmibawah251)
astat3, crit3, sig3 = stats.anderson(bmiatas251, dist='norm')
astat4, crit4, sig4 = stats.anderson(bmibawah251, dist='norm')
#print results
blah3 = astat3/crit3[4]
blah4 = astat4/crit4[4]
print('Shapiro-Wilks test')
print('Charges smoker:      p-value: %.8f' % pval3)
print('Charges non smoker:  p-value: %.8f' % pval4)
print('\nAnderson-Darling Test')
print('Charges smoker:      Anderson-Darling statistic is %.2f times the critical value for alpha = 0.01' % blah3)
print('Charges non smoker:  Anderson-Darling statistic is %.2f times the critical value for alpha = 0.01' % blah4)
print('\nNumber of samples:')
print('Charges smoker:      n = {}' .format(len(bmiatas25)))
print('Charges non smoker:  n = {}' .format(len(bmibawah25)))
stat1, pval11 = stats.levene(bmiatas251, bmibawah251, center='mean')
if pval11 < 0.05:
    print('Tolak H0: Varians mempunyai nilai yang berbeda.\np-value: %.8f' % pval11)
else:
    print('Gagal tolak H0: H0 bernilai sama di semua sampel.\np-value: %.8f' % pval11)
statistic, pval = stats.ttest_ind(bmiatas251, bmibawah251, equal_var = False) 
if pval3 < 0.05:
    print('Tolak H0: charges BMI di atas 25 > BMI di bawah 25.\np-value: %.8f' % pval3)
else:
    print('Gagal tolak H0: Charges BMI di atas 25 dan charges BMI di bawah 25 bernilai sama.\np-value: %.8f' % pval4)


#3
print('\n\n\n3. BMI laki-laki dan perempuan sama:')
print('Uji Hipotesis: H0 = BMI male = BMI female. H1 = BMI male  BMI female.')
bmimale = df[(df.sex == 'male')]['bmi']
bmifemale = df[(df.sex == 'female')]['bmi']
ax8 = sns.distplot(bmimale, color='tab:blue')
ax8.axvline(np.mean(bmimale),color='red',linestyle='--', label='mean')
ax8.axvline(np.median(bmimale),color='green',linestyle='--', label='median')
plt.title('BMI Male')
plt.legend()
plt.show()
ax9 = sns.distplot(bmifemale, color='tab:pink')
ax9.axvline(np.mean(bmifemale),color='red',linestyle='--', label='mean')
ax9.axvline(np.median(bmifemale),color='green',linestyle='--', label='median')
plt.title('BMI Female')
plt.legend()
plt.show()
print('Skew of BMI male:          %.4f' % stats.skew(bmimale))
print('Skew of BMI female:      %.4f' % stats.skew(bmifemale))
print('Kurtosis of BMI male:     %.4f' % stats.kurtosis(bmimale))
print('Kurtosis of BMI female:  %.4f' % stats.kurtosis(bmifemale))
print('Dari uji tersebut terlihat bahwa distribusi keduanya mendekati distribusi normal.')
statistic5, pval5 = stats.shapiro(bmimale)
statistic6, pval6 = stats.shapiro(bmifemale)
astat5, crit5, sig5 = stats.anderson(bmimale, dist='norm')
astat6, crit6, sig6 = stats.anderson(bmifemale, dist='norm')
#print results
blah5 = astat5/crit5[4]
blah6 = astat6/crit6[4]
print('Shapiro-Wilks test')
print('BMI male:      p-value: %.8f' % pval5)
print('BMI female:  p-value: %.8f' % pval6)
print('\nAnderson-Darling Test')
print('BMI male:      Anderson-Darling statistic is %.2f times the critical value for alpha = 0.01' % blah5)
print('BMI female:  Anderson-Darling statistic is %.2f times the critical value for alpha = 0.01' % blah6)
print('\nNumber of samples:')
print('BMI male:      n = {}' .format(len(bmimale)))
print('BMI female:  n = {}' .format(len(bmifemale)))
stat2, pval22 = stats.bartlett(bmimale, bmifemale)
if pval22 < 0.05:
    print('Tolak H0: Varians mempunyai nilai yang berbeda.\np-value: %.8f' % pval22)
else:
    print('Gagal tolak H0: H0 bernilai sama di semua sampel.\np-value: %.8f' % pval22)
statistic, pval33 = stats.ttest_ind(bmimale, bmifemale, equal_var = True)
if pval33 < 0.05:
    print('Tolak H0: BMI male  BMI female.\np-value: %.8f' % pval3)
else:
    print('Gagal tolak H0: BMI male = BMI female.\np-value: %.8f' % pval4)
```
